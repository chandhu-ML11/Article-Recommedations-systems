```python

import pandas as pd
import numpy as np
from sklearn.feature_extraction import text
from sklearn.metrics.pairwise import cosine_similarity
data = pd.read_csv("articles (1).csv", encoding='latin1')
data.head()

```

```python
data.isnull().sum()
```

```python
print(list(data.columns))
```

```python
# 1. Use "title" because "Article" does not exist in your columns
articles = data["title"].tolist()

# 2. Initialize the Vectorizer (ensure input=articles is REMOVED from here)
uni_tfidf = text.TfidfVectorizer(stop_words="english")

# 3. Create the matrix
uni_matrix = uni_tfidf.fit_transform(articles)

# Check the result
print(f"Success! Matrix shape: {uni_matrix.shape}")
```

```python
uni_sim = cosine_similarity(uni_matrix)
```

```python
# 1. Calculate the Cosine Similarity Matrix first
uni_sim = cosine_similarity(uni_matrix)

# 2. Define the function (Using 'iloc' for safety and 'title' for correctness)
def recommend_articles(x):
    # argsort returns indices from lowest to highest similarity
    # [-5:-1] grabs the top 4 matches, excluding the article itself (which is always at -1)
    similar_indices = x.argsort()[-5:-1]

    # We use .iloc because argsort gives us the row NUMBER, not the index label
    return ", ".join(data["title"].iloc[similar_indices].values)

# 3. Apply the function
data["Recommended Articles"] = [recommend_articles(x) for x in uni_sim]

# 4. View the results
data.head()
```
